```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style = "whitegrid")
```

```python
num = np.random.randint(10, 200, 10)
num
```

```python
df = pd.DataFrame(num, columns=["value1"])
df
```

```python
df.describe()
```

```python
sns.boxplot(data=df, orient="v", width=0.2)
plt.show()
```

```python
sns.violinplot(data=df, orient="h", width=0.2)
```
